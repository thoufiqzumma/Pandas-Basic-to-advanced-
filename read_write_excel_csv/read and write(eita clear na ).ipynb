```python
#data read from excle 
def converts_people_value(cell):
    if cell=='n.a.':
        return ('walton mt')
    else:
        return cell

def converts_tickers_value(cell):
    if cell=='TATA':
        return ('DAFFODIL')
    else:
        return cell

df=pd.read_excel('stock_data.xlsx','Sheet1',converters={
    'people':converts_people_value,
    'tickers':converts_tickers_value
    
})
df
```
