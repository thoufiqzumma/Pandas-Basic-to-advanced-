```python
import pandas as pd
```

```python
df1=pd.DataFrame({'city':['dhaka','chattagram','khulna','barisal'],
                  'temp':[38,35,36,37],
                  'windspeed':[6,7,8,9]
    
})
df1
```

```python
df2=pd.DataFrame({'city':['new work','paris','london','istambul'],
                  'temp':[40,36,26,27],
                  'windspeed':[6,10,8,12]
    
})
df2
```

```python
df=pd.merge(df1,df2,on='windspeed',how='outer')#if ypu pass (how=outer) then show all element of the dataframe
df
```

```python
df=pd.marge(df1,df2,how='right')
df
```
